```python
import dowhy
from dowhy import CausalModel
import pandas as pd
```

```python
import networkx as nx
```

```python
df = pd.read_csv('/content/rad_data_analysis.csv')
```

```python
print(df.keys())
```

```python
from sklearn.ensemble import HistGradientBoostingClassifier
```

```python
graph= """digraph {
    age -> treatment;
    ecog_ps -> treatment;
    smoking_status_clean -> treatment;
    t -> treatment;
    n -> treatment;
    m -> treatment;
    sex -> treatment;
    pathology_group -> treatment;
    smoking_py_clean -> treatment;
    primary_site_group -> treatment;
    hpv_clean -> treatment;


    age -> time_os_days;
    ecog_ps -> time_os_days;
    t -> time_os_days;
    n -> time_os_days;
    m -> time_os_days;
    sex -> time_os_days;
    hpv_clean -> time_os_days;
    pathology_group -> time_os_days;
    smoking_py_clean -> time_os_days;
    primary_site_group -> time_os_days;



    treatment -> time_os_days;
}"""

# I. Create a causal model from the data and given graph.
model = CausalModel(
    data = df,
    treatment= "treatment",
    outcome= "time_os_days",
    graph= graph # Or alternatively, as nx.DiGraph
)

# II. Identify causal effect and return target estimands
identified_estimand = model.identify_effect()

# III. Estimate the target estimand using a statistical method.
estimate = model.estimate_effect(
    identified_estimand,
    method_name="backdoor.propensity_score_matching",
    method_params={
        "propensity_score_model": HistGradientBoostingClassifier()  # ✅ Instance, not string
    }
)
```

```python

print(estimate)
```
